```julia
function femp(x)
    y = zeros(length(x))
    for i in 1:length(x)
        for j in x
            (j <= x[i]) && (y[i] += 1)
        end
    end
    return y/length(x)
end
```

```julia
using PyPlot
x = rand(1000)
y = femp(x)
plot(x, y, ".");
xlabel("\$x\$");
ylabel("\$F_{emp}(x)\$");
title("empirische Verteilungsfunktion");
```

```julia
h = plt[:hist](x, 20);
title("Histogramm der Gleichverteilung");
```

```julia
using Distributions
d = Normal()
x = linspace(-5, 5, 1000)
y = Distributions.pdf.(d, x)
plot(x, y);
title("Dichtefunktion der Standardnormalverteilung");
```

```julia
x = linspace(-5, 5, 1000)
y = Distributions.cdf.(d, x)
plot(x,y);
title("Verteilungsfunktion der Standardnormalverteilung");
```

```julia
u = zeros(1000)
for i in 1:1000
    u[i] = sum(rand(500))
end
m = mean(u)
sm = sum((u - m).^2)
u = (u - m) 
u = u ./ sqrt(1/1000 * sm);
```

```julia
y = femp(u)
plot(u, y, ".");
title("empirische Verteilungsfunktion");
xlabel("\$u\$");
ylabel("\$F_{emp}(u)\$");
```

```julia
h = plt[:hist](u, 20);
title("Histogramm der Normalverteiltung");
```

Die Summen der gleichverteilten Zahlen sind normalverteilt. Dies ist durch den zentralen Grenzwertsatz zu begrnden, der (grob zusammengefasst) aussagt, dass sich die Summe von unabhngigen Zufallsvariablen aus der gleichen Verteilung mit steigender Summandenzahl einer Normalverteilung annhert.
