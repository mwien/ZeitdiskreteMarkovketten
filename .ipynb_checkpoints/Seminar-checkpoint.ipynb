```julia
using Plots;
Plots.plotly();
```

```julia
function plot_birth_death_process(b, d, N)
    # create the matrix D
    D = zeros(N+1, N+1);
    D[1,1] = 1;
    for i = 2:N
        D[i, i-1] = d[i];
        D[i, i] = - b[i] - d[i];
        D[i, i+1] = b[i];
    end
    D[N+1, N] = d[N+1];
    D[N+1, N+1] = -d[N+1];
    # create the vector c
    c = ones(N+1);
    c *= -1;
    c[1] = 0;
    # calculate the expected extinction time
    τ = D^(-1)*c;
    # plot result
    x = collect(0:N);
    Plots.plot(x,τ)
end
```

```julia
br = 0.025;
dr = 0.025;
N = 20;
b = collect(0:N)*br
d = collect(0:N)*dr
Plots.plot(collect(0:20), [b,d, b+d])
```

```julia
plot_birth_death_process(b, d, N)
```

```julia
r = 0.015;
K = 10;
N = 20;
b = [r*(i - i^2/(2*K)) for i = 0:2*K];
d = [r*i^2/(2*K) for i = 0:2*K];
Plots.plot(collect(0:20), [b,d, b+d])
```

```julia
plot_birth_death_process(b, d, N)
```

```julia
r = 0.015;
K = 10;
N = 20;
b = [r*i for i = 0:N-1];
append!(b, 0)
d = [r*i^2/K for i = 0:N];
Plots.plot(collect(0:20), [b,d, b+d])
```

```julia
plot_birth_death_process(b, d, N)
```

```julia
function plot_dist(b, d, N)
    X = zeros(1000, N+1)
    X[1,2] = 1
    for i = 2:1000
        for j = 0:N
            X[i,j+1] += X[i-1, j+1]*(1-d[j+1]-b[j+1])
            (j > 1) && (X[i,j+1] += X[i-1,j]*b[j])
            (j < N) && (X[i,j+1] += X[i-1,j+2]*d[j+2])
        end
    end
    Plots.surface(collect(0:20), collect(1:1000), X)
end
```

```julia
r = 0.015;
K = 10;
N = 20;
b = [r*(i - i^2/(2*K)) for i = 0:2*K];
d = [r*i^2/(2*K) for i = 0:2*K];
plot_dist(b,d,N)
```

```julia
r = 0.015;
K = 10;
N = 20;
b = [r*i for i = 0:N-1];
append!(b, 0)
d = [r*i^2/K for i = 0:N];
plot_dist(b,d,N)
```
